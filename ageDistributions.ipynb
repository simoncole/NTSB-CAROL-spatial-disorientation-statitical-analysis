```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
# Read in the datasets
all_fatal = pd.read_csv('../data/all_fatal.csv')
all_fatal_important = pd.read_csv('../data/all_fatal_important.csv')
spatial_disorientation_important = pd.read_csv('../data/spatial_disorientation_fatal_important.csv')
spatial_disorientation = pd.read_csv('../data/spatial_disorientation_fatal.csv')

all_ages = all_fatal['aircraft.Flight_Crew.crew_age'].dropna().to_numpy()
sd_ages = spatial_disorientation['aircraft.Flight_Crew.crew_age'].dropna().to_numpy()
```

```python
# Descriptive statistics for all_ages
print("All Fatal Crashes Age Statistics:")
print(f"Count: {len(all_ages)}")
print(f"Mean: {np.mean(all_ages):.2f}")
print(f"Std Dev: {np.std(all_ages):.2f}")
print(f"Min: {np.min(all_ages):.2f}")
print(f"25%: {np.percentile(all_ages, 25):.2f}")
print(f"Median: {np.median(all_ages):.2f}")
print(f"75%: {np.percentile(all_ages, 75):.2f}")
print(f"Max: {np.max(all_ages):.2f}")

print("\nSpatial Disorientation Age Statistics:")
print(f"Count: {len(sd_ages)}")
print(f"Mean: {np.mean(sd_ages):.2f}")
print(f"Std Dev: {np.std(sd_ages):.2f}")
print(f"Min: {np.min(sd_ages):.2f}")
print(f"25%: {np.percentile(sd_ages, 25):.2f}")
print(f"Median: {np.median(sd_ages):.2f}")
print(f"75%: {np.percentile(sd_ages, 75):.2f}")
print(f"Max: {np.max(sd_ages):.2f}")
```

```python
# Visualizations
try: 
    plt.figure(figsize=(12, 6))
    sns.histplot(data=all_ages, kde=True, label='All Fatal Crashes', color='blue', alpha=0.5)
    plt.legend()
    plt.title('Age Distribution Comparison')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.show()
except Exception as e:
    pass
```

```python
try: 
    # Visualizations
    plt.figure(figsize=(12, 6))
    sns.histplot(data=sd_ages, kde=True, label='Spatial Disorientation', color='red', alpha=0.5)
    plt.legend()
    plt.title('Age Distribution Comparison')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.show()
except Exception as e:
    pass
```

```python
# Create a histogram comparing the age distributions
plt.figure(figsize=(12, 6))

# Plot the histograms
sns.histplot(data=all_ages, kde=True, label='All Fatal Crashes', color='red', alpha=0.5, stat='density')
sns.histplot(data=sd_ages, kde=True, label='Spatial Disorientation Crashes', color='blue', alpha=0.5, stat='density')
# plt.hist(all_ages, bins=30, alpha=0.5, label='All Fatal Crashes', color='blue', density=True)
# plt.hist(sd_ages, bins=30, alpha=0.5, label='Spatial Disorientation', color='red', density=True)

plt.title('Age Distribution: All Fatal Crashes vs Spatial Disorientation Density Plot')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()

# Set x-axis limits to cover the range of ages
plt.xlim(0, max(max(all_ages), max(sd_ages)) + 5)

plt.show()
```

```python
# Normality test
print(stats.shapiro(all_ages))
print(stats.shapiro(sd_ages))

# T-test or Mann-Whitney U test
# t_stat, p_value = stats.ttest_ind(all_ages, sd_ages)
# or
u_stat, p_value = stats.mannwhitneyu(all_ages, sd_ages)

print(f"p-value: {p_value}")
```
