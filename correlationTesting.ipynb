```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'year': [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'set1': [217, 240, 210, 227, 254, 260, 318, 312, 379, 444, 473, 629, 801, 1147, 1217],
    'set2': [23.7, 3.98, 15.44, 34.7, 47.5, 77.73, 59.08, 56.01, 53.71, 65.94, 77.31, 118.89, 232.9, 351.67, 392.2]
})

print(data)

```

```python
# Convert DataFrame columns to numpy arrays
years = data['year'].to_numpy()
set1 = data['set1'].to_numpy()
set2 = data['set2'].to_numpy()

# Create a line plot overlaying set1 and set2 over time
plt.figure(figsize=(12, 6))

# Plot set1
plt.plot(years, set1, label='Set 1', marker='o')

# Plot set2
plt.plot(years, set2, label='Set 2', marker='s')

plt.title('Set 1 and Set 2 over Time')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate Pearson's correlation coefficient between set1 and set2
correlation_coefficient = np.corrcoef(data['set1'], data['set2'])[0, 1]

print(f"Pearson's correlation coefficient between set1 and set2: {correlation_coefficient:.4f}")
```

There is a very strong correlation between the two datasets

```python
# Create a scatter plot with Set 1 on x-axis and Set 2 on y-axis
plt.figure(figsize=(12, 6))
plt.scatter(set1, set2, alpha=0.7)

# Add trend line
z = np.polyfit(set1, set2, 1)
p = np.poly1d(z)
plt.plot(set1, p(set1), "r--", alpha=0.7)

plt.title('Scatter Plot: Set 1 vs Set 2 with Trend Line')
plt.xlabel('Data Set 1')
plt.ylabel('Data Set 2')
plt.grid(True)

# Show the plot
plt.show()

# Print the equation of the trend line
print(f"Trend line equation: y = {z[0]:.4f}x + {z[1]:.4f}")
```
